```python
def pad_time(x,n):
    if len(x) != n:
        x=x.rjust(n,'0')
    return x

#x=x[0:2]+':'+x[2:4]
```

```python
df_init.an = df_init.an.astype('str').apply(pad_time, args=(2,))
df_init.mois = df_init.mois.astype('str').apply(pad_time, args=(2,))
df_init.jour = df_init.jour.astype('str').apply(pad_time, args=(2,))
df_init.hrmn= df_init.hrmn.astype('str').apply(pad_time, args=(4,))
#df_init.hrmn= df_init.hrmn[0:2]+'-'+df_init.hrmn[2:4]
```

## Is the number of accidents decreasing or increasing?

```python
df_init.columns
```

```python
df_init['hr']= df_init['hrmn'].astype('str').str.slice(-4,-2)
```

```python
init_notebook_mode(connected=True)
topcolor = 'rgb(128, 0, 128)'
othercolor = 'rgb(128, 128, 0)'
def plotbar(data, n):
    topn = data.sort_values(ascending=False)[0:n]
    colors = [topcolor if i in topn else othercolor for i in data.index]
    bar = go.Bar(x=data.index,y=data.values, marker=dict(color = colors))
    return bar

by_year = df_init.groupby(['an']).size()
by_month = df_init.groupby(['mois']).size()
by_day = df_init.groupby(['jour']).size()
by_hour = df_init.groupby(['hr']).size()

figure =tools.make_subplots(rows=2, cols=2, \
                            subplot_titles=('By Year', 'By Month', 'By Day', 'By Hour'))
trc1 = plotbar(by_year, 3)
trc2 = plotbar(by_month, 3)
trc3 = plotbar(by_day, 5)
trc4 = plotbar(by_hour, 5)

figure.append_trace(trc1, 1,1)
figure.append_trace(trc2, 1,2)
figure.append_trace(trc3, 2,1)
figure.append_trace(trc4, 2,2)

figure['layout'].update(height=900, width=900, title='Accidents Grouped With Timelines')
iplot(figure,filename='make-subplots-multiple-with-titles')
```
